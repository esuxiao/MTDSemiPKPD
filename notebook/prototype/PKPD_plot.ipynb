```r
rm(list = ls())

source_all_files <- function(path){
  files <- list.files(path)
  for(i in 1:length(files)){
    full_path <- paste(path, files[i],sep="")
    source(full_path)
  }
}
### Load the new version
library(ggplot2)
library(MTDSemiPKPD)
new_verrsion_path <- 'C:/Users/Xiao/Documents/Xiao_folder/GitHub/MTDSemiPKPD/MTDSemiPKPD/R/'
source_all_files(new_verrsion_path)
```

```r
k10 <- 0.5
Vc <- 2
Emax <- 1.25
ED50 <- 2
gamma <- 0.5

pkpd_model_obj <- PKPD_model.create(
    administration_dosage = c(1,1,1, 1),
    administration_timepoint =  c(0, 5, 14, 21),
    PK_model_type = 'IV',
    PK_model_parms = c(k10, Vc),
    PD_model_type = 'Emax',
    PD_model_parms = c(Emax, ED50, gamma)
    )
```

### PK curve: concentration vs time

```r
PKPDmodel.plot_PK_curve(pkpd_model_obj, seq(from = 0, to =20, by = 0.1))
```

### PD curve: effect vs concentration

```r
PKPDmodel.plot_PD_curve(pkpd_model_obj, seq(from = 0, to =1, by = 0.01))
```
